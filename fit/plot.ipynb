```python
import sys,os
import numpy as np
import pylab as py
import pandas as pd
%matplotlib inline
from  matplotlib import rc
import matplotlib.pyplot as plt
from matplotlib.colors import LogNorm
rc('font',**{'family':'sans-serif','sans-serif':['Helvetica']})
rc('text',usetex=True)
```

## F1

```python
f1s = {}
f1s['lattice'] = np.loadtxt('f1smod.txt', skiprows = 1)
f1s['lfhqcd original'] = np.loadtxt('results/f1s_lfhqcd_original.dat', skiprows = 1)
f1s['lfhqcd fit'] = np.loadtxt('results/f1s_lfhqcd.dat', skiprows = 1)
f1s['lfhqcdmass original'] = np.loadtxt('results/f1s_lfhqcdmass_original.dat', skiprows = 1)
f1s['lfhqcdmass fit'] = np.loadtxt('results/f1s_lfhqcdmass.dat', skiprows = 1)

```

```python
nrows, ncols = 1, 1

fig = plt.figure(figsize=(ncols*7, nrows*7*0.75))

ax = fig.add_subplot(nrows, ncols, 1)
c0 = ax.errorbar(f1s['lattice'][:,0], f1s['lattice'][:,1], yerr = f1s['lattice'][:,2],\
            marker = 'o', markersize = 8, color = 'blue', linestyle = '', capsize = 5)

c1 = ax.fill_between(f1s['lfhqcd fit'][:,0], f1s['lfhqcd fit'][:,1] - f1s['lfhqcd fit'][:,2], f1s['lfhqcd fit'][:,1] + f1s['lfhqcd fit'][:,2],\
               color = 'red', alpha = 0.85, linewidth = 0)
c2 = ax.plot(f1s['lfhqcd fit'][:,0], f1s['lfhqcd fit'][:,1],\
        color = 'darkred', linestyle = '--', linewidth = 2)

c3 = ax.fill_between(f1s['lfhqcdmass fit'][:,0], f1s['lfhqcdmass fit'][:,1] - f1s['lfhqcdmass fit'][:,2], f1s['lfhqcdmass fit'][:,1] + f1s['lfhqcdmass fit'][:,2],\
               edgecolor = 'darkturquoise', alpha = 0.85, linewidth = 2, facecolor = '')
c4 = ax.plot(f1s['lfhqcdmass fit'][:,0], f1s['lfhqcdmass fit'][:,1],\
        color = 'cyan', linestyle = '-.', linewidth = 2)
ax.set_xlabel(r'$Q^2~(\rm GeV^2)$', size = 25)
ax.set_ylabel(r'$F_1^s(Q^2)$', size = 25)
ax.set_xlim(0, 1.5)
ax.set_ylim(0, 0.005)
ax.tick_params(axis = 'both', which = 'both', direction = 'in', labelsize = 20)
L = [c0, (c1, c2[0]), (c3, c4[0])]
H = [r'$\rm Lattice$', r'$\rm LFHQCD(massless)$', r'$\rm LFHQCD(massive)$']
ax.legend(L, H, frameon = False, loc = 'upper left', fontsize = 18)


fig.tight_layout()
fig.savefig('results/compare_f1s_1.pdf')
```

```python
nrows, ncols = 1, 1

fig = plt.figure(figsize=(ncols*7, nrows*7*0.75))

ax = fig.add_subplot(nrows, ncols, 1)
c0 = ax.errorbar(f1s['lattice'][:,0], f1s['lattice'][:,1], yerr = f1s['lattice'][:,2],\
            marker = 'o', markersize = 8, color = 'blue', linestyle = '', capsize = 5)

c2 = ax.plot(f1s['lfhqcd original'][:,0], f1s['lfhqcd original'][:,1] * 4.88,\
        color = 'darkred', linestyle = '--', linewidth = 2)

c4 = ax.plot(f1s['lfhqcdmass original'][:,0], f1s['lfhqcdmass original'][:,1] * 4.88,\
        color = 'cyan', linestyle = '-.', linewidth = 2)
ax.set_xlabel(r'$Q^2~(\rm GeV^2)$', size = 25)
ax.set_ylabel(r'$F_1^s(Q^2)$', size = 25)
ax.set_xlim(0, 1.5)
ax.set_ylim(0, 0.005)
ax.tick_params(axis = 'both', which = 'both', direction = 'in', labelsize = 20)
L = [c0, (c2[0]), (c4[0])]
H = [r'$\rm Lattice$', r'$\rm LFHQCD(massless)$', r'$\rm LFHQCD(massive)$']
ax.legend(L, H, frameon = False, loc = 'upper left', fontsize = 18)


fig.tight_layout()
fig.savefig('results/compare_f1s_2.pdf')
```

## q(x)

```python
qs = {}

#qs['lfhqcd original'] = np.loadtxt('results/ssbar_lfhqcd_original.dat', skiprows = 1)
qs['lfhqcd fit'] = np.loadtxt('results/ssbar_lfhqcd.dat', skiprows = 1)
#qs['lfhqcdmass original'] = np.loadtxt('results/ssbar_lfhqcdmass_original.dat', skiprows = 1)
qs['lfhqcdmass fit'] = np.loadtxt('results/ssbar_lfhqcdmass.dat', skiprows = 1)
```

```python
nrows, ncols = 1, 1

fig = plt.figure(figsize = (ncols * 7, nrows * 7 * 0.75))

ax = fig.add_subplot(nrows, ncols, 1)

c1 = ax.fill_between(qs['lfhqcd fit'][:,0],\
                qs['lfhqcd fit'][:,0] * (qs['lfhqcd fit'][:,1] - qs['lfhqcd fit'][:,2]),\
                qs['lfhqcd fit'][:,0] * (qs['lfhqcd fit'][:,1] + qs['lfhqcd fit'][:,2]),\
               color = 'red', alpha = 0.85, linewidth = 0)
c2 = ax.plot(qs['lfhqcd fit'][:,0], qs['lfhqcd fit'][:,0] * qs['lfhqcd fit'][:,1],\
       color = 'darkred', linestyle = '--', linewidth = 2)

c3 = ax.fill_between(qs['lfhqcdmass fit'][:,0],\
                qs['lfhqcdmass fit'][:,0] * (qs['lfhqcdmass fit'][:,1] - qs['lfhqcdmass fit'][:,2]),\
                qs['lfhqcdmass fit'][:,0] * (qs['lfhqcdmass fit'][:,1] + qs['lfhqcdmass fit'][:,2]),\
               edgecolor = 'cyan', alpha = 0.85, linewidth = 2, facecolor = '')
c4 = ax.plot(qs['lfhqcdmass fit'][:,0], qs['lfhqcdmass fit'][:,0] * qs['lfhqcdmass fit'][:,1],\
       color = 'cyan', linestyle = '-.', linewidth = 2)

ax.set_xlim(1e-3, 1)
ax.set_ylim(-0.007, 0.007)
ax.set_xscale('log')
ax.set_xlabel(r'$x$', size = 25)
ax.set_ylabel(r'$x[s(x)-\bar{s}(x)]$', size = 25)
ax.tick_params(axis = 'both', which = 'both', direction = 'in', labelsize = 20)

L = [(c1, c2[0]), (c3, c4[0])]
H = [r'$\rm LFHQCD(massless)$', r'$\rm LFHQCD(massive)$']
ax.legend(L, H, frameon = False, loc = 'upper left', fontsize = 18)

fig.tight_layout()
fig.savefig('results/compare_qs.pdf')
```
